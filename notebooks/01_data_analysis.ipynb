```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/dataset.csv")

print(f"Total examples: {len(df)}")
print(f"\nLabel distribution:\n{df['label'].value_counts()}")
print(f"\nMissing values:\n{df.isnull().sum()}")
```

```python

df['label'].value_counts().plot(kind='bar', color=['steelblue', 'tomato'])
plt.title('Label Distribution')
plt.xlabel('Label')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('../data/label_distribution.png')
plt.show()
```

```python

df['text_length'] = df['text'].apply(len)

print(df.groupby('label')['text_length'].describe())

df.groupby('label')['text_length'].plot(kind='hist', alpha=0.6, bins=30)
plt.title('Text Length Distribution by Label')
plt.xlabel('Character Count')
plt.ylabel('Frequency')
plt.legend(['safe', 'injection'])
plt.tight_layout()
plt.savefig('../data/text_length_distribution.png')
plt.show()
```
